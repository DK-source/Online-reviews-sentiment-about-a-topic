```python
rest_train = pd.read_csv('./Datasets/yelp_review_polarity_csv/train.csv')
rest_train
```

```python
review = []
for col in rest_train:
    if col == '1':
        pass
    else:
        review.append(col)
        for text in rest_train[col].tolist():
            review.append(text)

Bool = []
for col in rest_train:
    if col == '1':
        Bool.append(int(col))
        for text in rest_train[col].tolist():
            Bool.append(text)
```

```python
rest_train.loc[len(rest_train.index)] = [0,0]
```

```python
rest_train['review'] = review
rest_train['target'] = Bool
rest_train.drop(rest_train.columns[[0,1]], axis=1, inplace=True)
```

```python
rest_train.loc[rest_train['target'] == 1, 'target'] = 0 #Order Matter
rest_train.loc[rest_train['target'] == 2, 'target'] = 1
```

```python
rest_train
```

```python
print('Recommended reviews: ', len(rest_train[rest_train['target']==1]))
print('Not recommended reviews: ', len(rest_train[rest_train['target']==0]))
```

```python
rest_train.to_csv('./Datasets/yelp_review_polarity_csv/fixed_train.csv', index=False)
```
